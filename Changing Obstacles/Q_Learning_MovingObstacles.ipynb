```python
import numpy as np
import time
import random
import matplotlib.pyplot as plt
#Defining Grid

#defining action variables
up=0
down=1
left=2
right=3

#no. of states and no. of variables
noA=4
noC=16

C=range(noC)

step_reward=-1    #for every step
wall_reward=-10
out_of_grid_reward=-5
goal_reward=100

terminal_state = lambda s: s==noC-1  #last cell is terminal

def getPlayerPos(state):
    for i in range(16):
        if state[i]=="1":
            return i

def is_wall(state,x):
    if state[x]=="7":
        return True

def getNextState_Reward(state,action):
    state=list(state)
    player_pos=getPlayerPos(state)
    
    if terminal_state(player_pos):
        reward=100.0
        state="".join(state)
        return state,reward
    
    if action==up:
        player_new_pos=player_pos if(player_pos<4) else player_pos-4
    elif action==down:
        player_new_pos=player_pos if(16-player_pos<=4) else player_pos+4
    elif action==left:
        player_new_pos=player_pos if(player_pos%4==0) else player_pos-1
    elif action==right:
        player_new_pos=player_pos if((player_pos+1)%4==0) else player_pos+1    
    
    if is_wall(state,player_new_pos):
        reward=-10
        #print "wall"
        #print "Action that got here",action
        state="".join(state)
        return state,reward
    else:
        if terminal_state(player_new_pos): #reached termination
            reward=100
        elif player_new_pos==player_pos: #out of grid action
            reward=-5
        else:
            reward=-1
        state[player_pos]="0"
        state[player_new_pos]="1"
        state="".join(state)
        #print "Hey"
        #print "Player pos",player_pos,player_new_pos,"action",action
        #print "New State",state
        return state,reward

walls=[5,7,9]
curr_state=np.zeros((16,))
curr_state[walls]=5
```

```python
def e_greedy(e,Q_s):
    x=random.randrange(1,11)
    if x<=e*10:
        return random.randrange(noA)
    else:
        return np.argmax(Q_s)
    
```

```python
from scipy.stats import truncnorm

def get_truncated_normal(mean=0, sd=1, low=0, upp=10):
    return truncnorm(
        (low - mean) / sd, (upp - mean) / sd, loc=mean, scale=1)
```

```python
def q_learning(no_episodes,no_steps,alpha,discount,delta_eps):
    fails=0
    Q=dict()
    epsilon=1
    Rewards=[]
    States_Covered=[]
    
    for e in range(no_episodes):
        temp=get_truncated_normal(mean=7, sd=1, low=0, upp=15)
        walls=temp.rvs(3).astype(int)
       
        player_pos=random.randrange(16)
        while player_pos in walls:
            player_pos=random.randrange(16)
        
        start_state=np.zeros((16,)).astype(int)
        start_state[walls]=7
        start_state[player_pos]=1
        
        curr_state="".join(map(str,start_state))
        
        step=0
        Episode_Reward=0
        
        while terminal_state(getPlayerPos(curr_state))==False and step<no_steps:
            if curr_state not in Q.keys():
                Q[curr_state]=np.zeros((4,)).astype(int)
            A=e_greedy(epsilon,Q[curr_state])
            next_state,R=getNextState_Reward(curr_state,A)
            
            if next_state not in Q.keys():
                Q[next_state]=np.zeros((4,)).astype(int)
            
            A_=np.argmax(Q[next_state])
            Q[curr_state][A]=Q[curr_state][A]+alpha*(R+discount*Q[next_state][A_]-Q[curr_state][A])
            curr_state=next_state
            
            Episode_Reward+=R
            step+=1
            
        Rewards.append(Episode_Reward)
        States_Covered.append(len(Q.keys()))
        
        if terminal_state(getPlayerPos(curr_state))==False:
            fails+=1
            print "Episode no",e, "Didn't Converge"
        else:
            print "Episode no",e,"Steps for convergence",step
            
        epsilon -= (1 - delta_eps)/no_episodes
        
    print ""    
    print "No. of episodes that didn't converge",fails        
    return Q,Rewards,States_Covered

result_Q,Rewards,States_Covered=q_learning(20000,200,0.1,0.7,0.1) 
#print result_Q

print ""
print ""
print "Best Policy for each state/grid configuration"
print ""
best_policy=dict()
for state in result_Q:
    best_action=np.argmax(result_Q[state])
    temp=list(state)
    if temp[-1]=="1":
        best_action=5
    best_policy[state]=best_action
    print "State",state, "Best Action",best_action
    

print ""
print "Rewards"
plt.plot(Rewards)
plt.xlabel('No. of episodes')
plt.ylabel('Reward in that Episode')
plt.show()

print ""
print "States Covered"
plt.plot(States_Covered)
plt.xlabel('No. of episodes')
plt.ylabel('No. of states encountered')
plt.show() 
```
